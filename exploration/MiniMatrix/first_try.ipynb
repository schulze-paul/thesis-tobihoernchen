```python
import sys
import time
sys.path.append("../..")
from thesis.envs.minimatrix_firsttry import MiniMatrix
from stable_baselines3 import A2C, PPO
from sb3_contrib import RecurrentPPO
from stable_baselines3.common.utils import set_random_seed
seed = 42
set_random_seed(seed)
```

```python
models_dir = "../../models/MiniMatrix"
logdir = "../../logs/MiniMatrix"
fleetsize = 1
max_fleetsize = 1
run_name = f"A2C-{fleetsize}-{max_fleetsize}-{time.time()}-{seed}"
```

```python
env = MiniMatrix("../../envs/MiniMatrix_Firsttry.zip", max_seconds = 60*60, fleetsize = fleetsize, max_fleetsize=max_fleetsize)
```

```python
model =A2C(
    "MlpPolicy", 
    env, 
    tensorboard_log= logdir,
    device = "cuda",
    policy_kwargs=dict(net_arch = [256,128,64,64,32,64])
    )
```

```python
TIMESTEPS = 100000
for i in range(1, 3):
    model.learn(total_timesteps=TIMESTEPS, reset_num_timesteps=False, tb_log_name=run_name)
    model.save(f"{models_dir}/{run_name}-{TIMESTEPS * i}")
```
